```python
import torch
import torch.nn as nn
import os
import re
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from PIL import Image
def rename_file(path, label):
  i = 0

  for filename in os.listdir(path):
    try:
      f, extension = os.path.splitext(path + filename)
      src = path + filename
      dst = path + label + str(i) + extension

      if not os.path.exists(dst):
        os.rename(src, dst)
      else:
        print(f'File {dst} already exists, skipping.')

      i += 1

    except Exception as e:
      print(e)
      i += 1
path = "D:\FYP\Dust and power loss data set\Solar_panel_dataset\dusty"
label = "dusty"
rename_file(path,label)

def imageResize(srcpath, despath):
    for filename in os.listdir(srcpath):
        try:
            img = Image.open(srcpath + filename)
            newimg = img.resize((64, 64))
            if not os.path.exists(despath):
                os.makedirs(despath)
            newimg.save(despath + filename)
            print('Resized and saved successfully {}.'.format(filename))

        except Exception as e:
            print(e)
            continue

srcpath = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\dusty"
despath = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\dusty_train_set"
print("starting resizing for test set")
imageResize(srcpath, despath)
srcpath = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\random"
despath = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\test_set"
imageResize(srcpath, despath)

def getdata(path):
    all_images_as_array = []
    label = []

    for filename in os.listdir(path):
        try:
            if re.match(r'dusty', filename):
                label.append(1)
            else:
                label.append(0)

            img = Image.open(path + filename)
            np_array = np.asarray(img.convert('L'))

            all_images_as_array.append(np_array)
        except:
            continue
    return np.array(all_images_as_array), np.array(label).reshape(1, -1)



path_train = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\dusty"
train_set_x_orig, train_set_y = getdata(path_train)
print("train set created")
path_test = r"D:\FYP\Dust and power loss data set\Solar_panel_dataset\random"
test_set_x_orig, test_set_y = getdata(path_test)
print("test set created")
print('X_train set : ', train_set_x_orig)
print('y_train set : ', train_set_y)
print('X_test set : ', test_set_x_orig)
print('y_test set : ', test_set_y)
```
